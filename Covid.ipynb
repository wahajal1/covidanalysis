```python
#importing libraies and labeling them for easier calling
import pandas as pd  #reads csv, data in tabular form, analyzes and cleans and does operations on the data.
import matplotlib.pyplot as plt #matplotlib generates graphs and pyplot acts like a pen so it helps with drawing
import seaborn as sns #same as matplotlib but easier and better
```

```python
df = pd.read_csv("/content/covid_sample_data.csv") #reads the file and puts it in a variable
df.head() #shows first 5 rows by default to make sure it reads them correctly
```

```python
df.shape #gives us the tables form as in numbe of rows and columns
df.info() #info about the table as in number of null and nonnull values and the type of data
df.describe() #statistical information
```

```python
df_saudi = df[df['location'] == 'Saudi Arabia'] #this filters so it only does a table on Saudi Arabia, df = DataFrame which means table in python
plt.figure(figsize=(10, 6)) #the size of the graph (width,height)
plt.plot(df_saudi['date'], df_saudi['new_cases']) #(x-axis = date, y-axis = number of cases)
plt.title("new cases in Saudi Arabia") #the title of the graph
plt.xlabel("Date") #labeling the x-axis
plt.ylabel("Number of cases") #labeling the y-axis
plt.xticks(rotation=45) #rotating the x-axis's data
plt.show() #shows the final graph
```
